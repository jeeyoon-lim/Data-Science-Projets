```
#Reading Historic dataset
df_historic = pd.read_csv('NYPD_Complaint_Historic_Sample.csv')
```

```
df_historic.head()
```

```
#Reading the Current dataset

df_ytd = pd.read_csv('NYPD_Complaint_Data_Current__Year_To_Date_.csv')

df_ytd.head()
```

```
#Combining historic and current datasetes
complaint_data = pd.concat([df_historic,df_ytd], sort=True)

#complaint_data.iloc[99995:100010,0:4]
complaint_data
```

```
#Checking to see all the different Offense Descriptions in combined dataset
print(complaint_data.OFNS_DESC.unique())
```

```
#Dropping columns we don't need
complaint_data.drop(
    ['CMPLNT_NUM',
     'ADDR_PCT_CD',
     'RPT_DT',
     'PD_CD',
     'PD_DESC',
     'PREM_TYP_DESC',
     'PARKS_NM',
     'HADEVELOPT',
     'HOUSING_PSA',
     'X_COORD_CD',
     'Y_COORD_CD',
     'SUSP_AGE_GROUP',
     'SUSP_RACE',
     'SUSP_SEX',
     'TRANSIT_DISTRICT',
     'PATROL_BORO',
     'STATION_NAME',
     'VIC_AGE_GROUP',
     'VIC_RACE',
     'VIC_SEX',
     'CMPLNT_TO_DT',
     'CMPLNT_TO_TM',
     'CRM_ATPT_CPTD_CD',
     'JURISDICTION_CODE',
     'JURIS_DESC',
     'KY_CD',
     'LAW_CAT_CD',
     'LOC_OF_OCCUR_DESC',
     'Lat_Lon',
     ],
     axis=1, inplace=True)

complaint_data
```

```
#Renaming columns
complaint_data.rename({'CMPLNT_FR_DT': 'Date', 'CMPLNT_FR_TM': 'Time',
                       'BORO_NM':'Borough','CRM_ATPT_CPTD_CD':'Crime Status',
                        'LAW_CAT_CD':'Offense Level','LOC_OF_OCCUR_DESC':'Location Description',
                       'JURISDICTION_CODE': 'Jurisdiction Code','JURIS_DESC':'Jurisdiction Description',
                       'KY_CD':'Offense Classification Code', 'OFNS_DESC':'Offense Description',

                        }, axis=1, inplace=True)
```

```
complaint_data.head()
```

```
#Create a new column Timestamp by combining Date and Time 
import datetime
import numpy as np

complaint_data['Timestamp']=complaint_data['Date'] + ' ' + complaint_data['Time']

#complaint_data['Date'] = complaint_data['Timestamp']

#complaint_data.rename({'Date':'Timestamp'}, axis=1, inplace=True)

complaint_data
```

```
#Drop Time column
complaint_data.drop(['Time'],axis=1,inplace=True)
```

```
#Drop Date column
complaint_data.drop(['Date'],axis=1,inplace=True)
```

```
complaint_data.head()
```

```
#setting Timestamp type to datetime
complaint_data['Timestamp']= pd.to_datetime(complaint_data['Timestamp'],errors="coerce",format="%m/%d/%Y %H:%M:%S")
```

```
#Identifying outliers
#outliers = complaint_data[complaint_data['Timestamp'] < complaint_data['Timestamp']=='2006-01-01 00:00:00'].index
#print(outliers)
```

```
#Redefining dataframe to not include outliers by date
complaint_data = complaint_data[complaint_data['Timestamp'] >= '2006/01/01 00:00:00']
```

```
#set Timestamp as index
complaint_data = complaint_data.set_index('Timestamp')
```

```
complaint_data.head()
```

```
#sort by Timestamp index
complaint_data.sort_index(inplace=True)
```

```
complaint_data
```

```
#drop NA data entries
complaint_data.dropna(axis=0, inplace= True)
```

```
#drop data entries aren't Manhattan
complaint_data = complaint_data[complaint_data['Borough'] == 'MANHATTAN']
```

```
complaint_data.head()
```

```
#drop data entries with unneeded offense categories

complaint_data = complaint_data[complaint_data['Offense Description'] != 'LOITERING/GAMBLING (CARDS, DIC']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OFFENSES AGAINST MARRIAGE UNCL']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'FELONY SEX CRIMES']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'nan']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'ESCAPE 3']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'HOMICIDE-NEGLIGENT-VEHICLE']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'KIDNAPPING AND RELATED OFFENSES']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'UNLAWFUL POSS. WEAP. ON SCHOOL']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OTHER TRAFFIC INFRACTION']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OTHER STATE LAWS (NON PENAL LAW)']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'MISCELLANEOUS PENAL LAW']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OFFENSES AGAINST PUBLIC ADMINI']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'ADMINISTRATIVE CODE']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'AGRICULTURE & MRKTS LAW-UNCLASSIFIED']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OFFENSES AGAINST THE PERSON']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'NYS LAWS-UNCLASSIFIED FELONY']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OTHER STATE LAWS (NON PENAL LA']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OFFENSES RELATED TO CHILDREN']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'NYS LAWS-UNCLASSIFIED VIOLATION']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'OTHER STATE LAWS']
complaint_data = complaint_data[complaint_data['Offense Description'] != 'ANTICIPATORY OFFENSES']
```
